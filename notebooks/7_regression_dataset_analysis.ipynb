```python
# correcting data

values = {
    'vehicleType': 'limousine',
    'gearbox':'manuell',
    'model':'golf',
    'fuelType':'benzin',
    'notRepairedDamage': 'nein',
}

values
```

```python
dataset = dataset.fillna(value=values)
```

```python
X = dataset.iloc[:, 1:12].values # numpy format 

X.shape, type(X)
```

```python
X
```

```python
y = dataset.iloc[:, 0].values
y
```

```python
# preprocessing
#onehotencoder = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [0, 1, 3, 5, 8, 9, 10])],remainder='passthrough')
onehotencoder = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [0, 1, 3, 5, 8, 9, 10])], remainder='passthrough')
```

```python
X = onehotencoder.fit_transform(X)
X
```

```python
dataset.shape, X.shape
```

```python
# initial definition of neuron units

# (input + output)/2

(316 + 1)/2
```

```python
# building the neural network

regressor = Sequential([
    tf.keras.layers.InputLayer(shape=(316, )),
    tf.keras.layers.Dense(units=158, activation='relu'),
    tf.keras.layers.Dense(units=158, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='linear')
])
```

```python
regressor.summary()
```

```python
regressor.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
regressor.fit(X, y, batch_size=300, epochs=100)
```

```python
# In this case, we are not using a test database 

predict = regressor.predict(X)
predict
```
