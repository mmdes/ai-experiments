### Saving the classifier and classifying single plant 
The goal of this task is to train the model using the best parameters found in the previous challenge (/notebooks/27_CHALLENGE_2.ipynb), generate the final classifier, save the neural network structure to the disk in json format along with its weights in h5 format, load it and classify a register

```python
import pandas as pd
import tensorflow as tf
import sklearn
import scikeras
```

```python
pd.__version__, tf.__version__, sklearn.__version__, scikeras.__version__
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import utils as np_utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Dropout
from tensorflow.keras import backend as k
from tensorflow.keras.losses import KLDivergence
from tensorflow.keras.models import model_from_json
import numpy as np
```

```python
data = pd.read_csv('../data/iris/iris.csv')
data
```

```python
X = data.iloc[:,0:4]
X
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)
X
```

```python
y = data.iloc[:, 4]
y
```

```python
# Creating labels

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Now our output have size equal to 3 
y = np_utils.to_categorical(y)
y
```

```python
neural_network = Sequential([
    InputLayer(shape = (4,)),
    Dense(units=8 , activation = 'tanh', kernel_initializer='random_uniform'),
    Dropout(rate=0.2), # to avoid overfitting
    Dense(units=8, activation = 'tanh', kernel_initializer='random_uniform'),
    Dropout(rate=0.2),
    Dense(units=3, activation='softmax')
])

neural_network.compile(optimizer='adam', loss=KLDivergence(), metrics=['categorical_accuracy'])
neural_network.fit(X, y, epochs=100, batch_size=10)
```

```python
# Saving model's structure
model_json = neural_network.to_json()

with open('../data/challenge3/my_model.json', "w") as json_file:
    json_file.write(model_json)
```

```python
# Saving model's defined weights

neural_network.save_weights('../data/challenge3/my_model.weights.h5')
```

```python
# Destroying model from memory
neural_network = None
```

```python
# Loading model and its weights back

# Model structure
with open('../data/challenge3/my_model.json', "r") as json_file:
    loaded_json_model = json_file.read()
loaded_model = model_from_json(loaded_json_model)

# Model weights
loaded_model.load_weights("../data/challenge3/my_model.weights.h5")

# Compiling model 
loaded_model.compile(optimizer='adam', loss='kullback_leibler_divergence', metrics=['categorical_accuracy'])
```

```python
# Predicting a single register
sample = X[0].reshape(1,4)
predicted_y = loaded_model.predict(sample)
```

```python
# Getting the original label 
predicted_y = [np.argmax(t) for t in predicted_y]
original_label = label_encoder.inverse_transform(predicted_y)
original_label[0]
```

```python
# Creating and predicting another register from random data

random_sample = np.random.rand(4)

random_sample =random_sample.reshape(1,4)
predicted_random = loaded_model.predict(random_sample)

predicted_random = [np.argmax(t) for t in predicted_random]
print(f'This is the label for the random sample: {label_encoder.inverse_transform(predicted_random)[0]}')

```
