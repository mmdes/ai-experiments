## Multiclass Classification

Iris dataset: https://archive.ics.uci.edu/dataset/53/iris

```python
import pandas as pd
import numpy as np 
import tensorflow as tf
import sklearn
```

```python
pd.__version__, np.__version__, tf.__version__, sklearn.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv('../data/iris/iris.csv')
base
```

```python
X = base.iloc[:,0:4].values
X

```

```python
y = base.iloc[:,4].values
y
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
y = np_utils.to_categorical(y)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
X.shape, y.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# (predictor_veriables + outputs number) / 2 = units
(4 + 3)/2 # I rounded the result to 4 
```

```python
# building the model 
neural_network = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),
    tf.keras.layers.Dense(units=4, activation='relu'),
    tf.keras.layers.Dense(units=4, activation='relu'),
    tf.keras.layers.Dense(units=3, activation='softmax')
])
```

```python
neural_network.summary()
```

```python
neural_network.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
neural_network.fit(X_train, y_train, batch_size=10, epochs=1000)
```

```python
neural_network.evaluate(X_test, y_test)
```

```python
predict = neural_network.predict(X_test)
predict
```

```python
predict = predict > 0.5
predict
```

```python
y_test
```

```python
y_test2 = [np.argmax(t) for t in y_test]
print(y_test2)
```

```python
predict2 = [np.argmax(t) for t in predict]
print(predict2)
```

```python
from sklearn.metrics import accuracy_score

# accuracy
accuracy_score(y_test2, predict2)
```

```python
# 0 setosa, 1 versicolor, 2 virginica
confusion_matrix(y_test2, predict2)
```
